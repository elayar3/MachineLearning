```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/AIML/MLCass/week1/Salary_Data.csv')
df.head()
```

```python
df.describe()
```

```python
plt.title('Salary Distribution List')
sns.distplot(df['Salary'])
plt.show()
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],color='blue')
plt.title('Salary vs Experiance')
plt.xlabel(df['Salary'])
plt.ylabel(df['YearsExperience'])
plt.box(False)
plt.show()
```

```python
X=df.iloc[:, :1]
y=df.iloc[:, 1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
regressor=LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color='lightcoral')
plt.plot(X_train, y_pred_train, color='firebrick')
plt.title('Salary vs Experiance Train')
plt.ylabel(df['Salary'])
plt.xlabel(df['YearsExperience'])
plt.legend(['X_train/Pred(y_test)','X_train/y_train'], title='Sal/Exp', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color='lightcoral')
plt.plot(X_train, y_pred_train, color='firebrick')
plt.title('Salary vs Years of Experience(Training set)')
plt.xlabel(df['YearsExperience'])
plt.ylabel(df['Salary'])
plt.legend(['X_train/pred(y_test)', 'X_train/y_train'],title='Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
